Se tiene la función y = a * log(b) + b * c^2. Calcular las derivadas de y respecto a los
parámetros a, b, c, evaluando en los valores a=2, b=10, c=0.5. Se deben calcular las
derivadas por los siguientes métodos:

1) Derivadas analíticas (symbolic differentiation).

dy/da = log(b) = log(10) = 1

dy/db = a * 1/(b*ln(10)) + c^2 = 2 * 1/(10 * ln(10)) + 0.5^2 = 2 * 1/23,026 + 0.25 = 0.336

dy/dc = 2 * b * c = 2 * 10 * 0.5 = 10

2) Límite con un delta numérico pequeño (numerical differentiation).

```python
import numpy as np

def der(a,b,c):
  return (a * np.log10(b) + b * c**2)

s = 0.00001
a = 2
b = 10
c = 0.5

der_a = (der(a+s,b,c)-der(a,b,c)) / s
der_b = (der(a,b+s,c)-der(a,b,c)) / s
der_c = (der(a,b,c+s)-der(a,b,c)) / s

print('Derivada respecto de a = ', der_a)
print('Derivada respecto de b = ', der_b)
print('Derivada respecto de c = ', der_c)
```

3) Grafos y la regla de la cadena (chain rule differentiation).

```python
d = a * np.log10(b)
e = b * c**2
y = d + e

dy_dd = 1
dy_de = 1
dd_da = np.log10(b)
dd_db = a / (b*np.log(10))
de_db = c**2
de_dc = 2 * b * c

dy_da = dy_dd * dd_da
dy_db = dy_dd * dd_db + dy_de * de_db
dy_dc = dy_de * de_dc

print('Derivada respecto de a = ', dy_da)
print('Derivada respecto de b = ', dy_db)
print('Derivada respecto de c = ', dy_dc)
```

4) PyTorch (automatic differentiation).

```python
import torch
import numpy as np

a_t = torch.tensor(2., requires_grad = True)
b_t = torch.tensor(10., requires_grad = True)
c_t = torch.tensor(0.5, requires_grad = True)

y = a_t * torch.log10(b_t) + b_t * c_t**2
y.backward()

print("Dervative of the equation at a = 2 is: ", a_t.grad.detach().numpy())
print("Dervative of the equation at b = 10 is: ", b_t.grad.detach().numpy())
print("Dervative of the equation at c = 0.5 is: ", c_t.grad.detach().numpy())
```
